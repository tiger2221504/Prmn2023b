# 課題

```python
from sklearn import datasets

datasets = datasets.load_diabetes()

data = pd.DataFrame(data=datasets.data, columns=datasets.feature_names)
target = pd.DataFrame(data=datasets.target, columns=['disease progression'])

#ageを相関係数の高いs5に変更
X_train, X_test, y_train, y_test = train_test_split(data.s5.to_numpy().reshape(-1,1), target, test_size=0.3, random_state=0)

reg = LinearRegression()
reg.fit(X_train,y_train)
reg.score(X_test,y_test)
```
